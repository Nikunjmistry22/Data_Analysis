```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
```

# Data

```python
data=pd.read_csv('gas_prices.csv')
data
```

# Line Plot

```python
plt.figure(figsize=(8,5))
plt.title('Gas Price in different countries')
plt.plot(data.Year,data.Australia,'g.-',label="Aus")
plt.plot(data.Year,data.USA,'r.-',label='USA')
plt.plot(data.Year,data['UK'],'b.-',label='UK')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Prices')

plt.show()
```

```python
plt.title('Prices in Countries',fontdict={'fontweight':'bold'})
for i in data:
  if i!='Year':
    plt.plot(data.Year,data[i])

plt.xlabel('Years')
plt.ylabel('Prices')
plt.show()
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

#Histogram

```python
bins=np.arange(0,101,10)
plt.hist(fifa['Overall'],bins=bins,facecolor="green")
plt.xticks(bins)
plt.xlabel('No of players')
plt.ylabel('Skills')
plt.title('Distribution of players',fontdict={'fontweight':'bold','fontsize':'10'})
plt.show()
```

#Pie Chart

```python
fifa['Preferred Foot']
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
plt.pie([left,right],labels=['left','right'],autopct='%.2f %%',colors=['orange','yellow'])
plt.title('Preferred foot count')
plt.show()
```

#Bar

```python
data=fifa.head(10)
plt.figure(figsize=(8,8))
plt.bar(data.Name,data.Age,facecolor="violet")
plt.title('Football player\'s Age')
plt.ylabel('Age')
plt.xlabel('Name')
plt.show()
```

#Boxplot

```python
plt.style.use('default')
barcelona=fifa.loc[fifa['Club']=="FC Barcelona"]['Overall']
Madrid=fifa.loc[fifa['Club']=="Real Madrid"]['Overall']
plt.boxplot([barcelona,Madrid],labels=['FC Barcelona','Real Madrid'])
plt.ylabel('Overall')
plt.title('Comparison between Football Teams')
plt.show()
```

#Stem plot

```python
data1=fifa.sample(20)
plt.stem(data1['Name'],data1['Jersey Number'])
plt.xlabel('Names')
plt.ylabel('Jersey No')
plt.title('PLayer Jersey No')
plt.show()
```
